```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

#plt.figure(1)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo')

plt.subplot(222)  #rows columns grid
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.subplot(223)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.subplot(224)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

# Bar chart

```python
#vertical
x=['a','b','c','d','e']
xpos=np.arange(len(x))
plt.xticks(xpos,x)

y=[20,40,10,50,60]
a=[10,50,20,80,30]
plt.bar(xpos-0.2,y,width=.4,label="abcd")
plt.bar(xpos+0.2,a,width=.4,label="xyz")
plt.legend()
```

```python
#horizontal
x=['a','b','c','d','e']
ypos=np.arange(len(x))
plt.yticks(ypos,x)

y=[20,40,10,50,60]
a=[10,50,20,80,30]
plt.barh(ypos-.2,y,height=.4, label="abcd")
plt.barh(ypos+.2,a,height=.4,label="xyz")
plt.legend()
```

# histogram

```python
y=[20,40,10,50,60]
a=[10,50,20,80,30]
plt.hist([y,a],bins=[10,30,50,100],rwidth=.4,label=["abcd","xyz"])

plt.legend()
```

# PIE Chart

```python
y=[20,40,10,50,60]
a=['a','b','c','d','e']
plt.axis("equal")
plt.pie(y,labels=a,radius=2,shadow=True,autopct="%0.f%%",explode=[0,0.5,0,0,0],colors=["pink","red","green","yellow","blue"])
plt.show()
```

```python
y=[20,40,10,50,60]
a=['a','b','c','d','e']
plt.axis("equal")
plt.pie(y,labels=a,radius=2,shadow=True,autopct="%0.f%%",explode=[0,0.5,0,0,0],colors=["pink","red","green","yellow","blue"])
plt.savefig("1.png",bbox_inches="tight")
```
